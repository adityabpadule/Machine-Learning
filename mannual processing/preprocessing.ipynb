### Replace the missing values

### Function called as df.fillna(value)

```python
def replace(df):
    for i in df.columns:
        if i in cat:
            md=df[i].mode()[0]
            df[i]=df[i].fillna(md)
        else:
            mn=df[i].mean()
            df[i]=df[i].fillna(mn)
    print("Missing values replaced")
```

```python
replace(df)
```

### Seperate X and Y(Weight)

### Perform preprocessing on X

```python
X_cat = X.select_dtypes(include='object')
X_cat
```

```python
df['Cylinders'].unique()
```

```python
df['Cylinders'].value_counts()
```

```python
X_con = X.select_dtypes(exclude='object')
X_con
```

### Apply one hot ecoding on categorical features

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore',
                    sparse_output=False).set_output(transform='pandas')
X_cat_pre = ohe.fit_transform(X_cat)
X_cat_pre.head()
```

```python
X_cat_pre.to_csv("ohe.csv", index=False)
```

Apply standard scaler on numerical features

```python
from sklearn.preprocessing import StandardScaler
```

```python

scaler = StandardScaler().set_output(transform='pandas')
X_con_pre = scaler.fit_transform(X_con)
X_con_pre.head()
```

```python
X_con_pre.shape
```

### Combine both categorical and continous preprocessed data

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

```python
X_pre.to_csv("pre.csv", index=False)
```

### Apply Train Test split

![Screenshot 2024-05-22 184527.png](<attachment:Screenshot 2024-05-22 184527.png>)

```python

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

### Preprocessing is completed
Build the model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

Evaluate the model

```python
# Get r2 score for train
model.score(xtrain, ytrain)
```

```python
# Get r2 score for test
model.score(xtest, ytest)
```
